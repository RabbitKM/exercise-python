```python
#JSON-圖片
import requests
import io
import json
import codecs

fn = 0
for p in range(1,6,1):
  req = requests.get(
      'https://udn.com/api/more',
      headers={
          'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36',
          'Referer': 'https://data.gov.tw/datasets/search?p=1&size=10&rct=285&rafft=6410',
          'Cookie': '_ga=GA1.3.253445803.1652016416; auth.strategy=local; _gid=GA1.3.1823959815.1652333736',
      },
      params={
          'page': p,
          'id': '',
          'channelId': '1',
          'cate_id': '0',
          'type': 'breaknews',
          'totalRecNo': '11260',

      },
      data={},
  )
  ret = json.loads(req.text)

  for dd in ret['lists']:
    req2 = requests.get(dd['url'])
    fn += 1
    with codecs.open('news_image/' + str(fn) + '.jpg', 'wb') as f:
      f.write(req2.content)
    print(dd['time']['date'],dd['title'])
```
