### 1. 讀取網址資料

```python
# https://www.104.com.tw/jobs/search/list?ro=0&kwop=7&keyword=python&expansionType=area,spec,com,job,wf,wktm&area=6001001000,6001003000&order=14&asc=0&page=2&mode=s&jobsource=n_my104_search&langFlag=0&langStatus=0&recommendJob=1&hotJob=1
import requests
req = requests.get(
    "https://www.104.com.tw/jobs/search/list",
    headers={
      "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36",
      "Referer": "https://www.104.com.tw/jobs/search/?ro=0&keyword=python&expansionType=area,spec,com,job,wf,wktm&area=6001001000,6001003000&order=14&asc=0&page=1&mode=s&jobsource=n_my104_search&langFlag=0&langStatus=0&recommendJob=1&hotJob=1",
      # "Cookie": "",
    },
    params={
      "ro": "0",
      "kwop": "7",
      "keyword": "python",
      "expansionType": "area,spec,com,job,wf,wktm",
      "area": "6001001000,6001003000",
      "order": "14",
      "asc": "0",
      "page": "1",
      "mode": "s",
      "jobsource": "n_my104_search",
      "langFlag": "0",
      "langStatus": "0",
      "recommendJob": "1",
      "hotJob": "1",
    }
)

req
```

```python
# Test: 各種方法
print("1. 讀取的狀態:\n", req.status_code)
print("2. 讀取的標頭資訊(字典類型):\n", req.headers)
print("3. 讀取和設定編碼:\n", req.encoding)
print("4. 讀取內容(以純文字型態讀取，會自動做編碼的轉換):\n", req.text)
print("5. 讀取內容(原始型態，可讀取二進位資料):\n", req.content)
print("6. 讀取內容(JSON資料):\n", req.json()) # 若req無設定headers則無法讀取
```

### 2. 找到內文位置

```python
text_dict = req.json()
for k, v in text_dict["data"]["list"][0].items():
   print(k, ":", v)
```

### 3. 將內文轉成表格

```python
import pandas as pd
# job_df = pd.DataFrame.from_dict(text_dict["data"]["list"]) # 不需使用from_dict, 原本就可直接放入字典使用.
job_df = pd.DataFrame(text_dict["data"]["list"])
job_df
```

```python
# Test1: 使用迴圈+append逐筆加入
test_df = pd.DataFrame()
for i in text_dict["data"]["list"]:
  test_df = test_df.append(i, ignore_index=True) # 不加ignore_index=True會跳ERROR
test_df
```
