```python
# 抓取影片的前景輪廓 (步驟3: 儲存影片測試2 - 只顯示最寬的框)
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time

p1 = cv2.VideoCapture('h3.mp4')
w = int(p1.get(3)) # 寬
h = int(p1.get(4)) # 高
f = int(p1.get(5)) # 每秒的影格數


p2 = cv2.VideoWriter('09-hw3.mp4', cv2.VideoWriter_fourcc(*'MP4V'), f, (w,h))
while p1.isOpened()==True:
  ret, m1 = p1.read()
  if ret == True:
    dp.clear_output(True)

    # 開始處理影像
    m2 = cv2.cvtColor(m1,cv2.COLOR_BGR2HSV) # 轉HSV
    m2 = cv2.inRange(m2[:,:,1], (130), (255)) # 取S範圍
    m2 = cv2.dilate(m2, np.ones((25,10))) # 調整膨脹大小, 減少輪廓數量.

    a, b = cv2.findContours(m2, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE) # 取輪廓點

    # 有輪廓再繪製矩形  
    if len(a) > 0:
      # 取輪廓最寬的點
      maxW = 0
      for i in range(1,len(a),1):
        x1, y1, w1, h1 = cv2.boundingRect(a[i-1])
        x2, y2, w2, h2 = cv2.boundingRect(a[i])
        if w2 > w1:
          maxW = i

      x, y, w, h = cv2.boundingRect(a[maxW]) # 取得包覆最小正矩形
      cv2.rectangle(m1, (x,y), (x+w,y+h), (0,0,255), 2) # 繪製矩形

    cv2_imshow(m1)
    
    # p2寫入影像
    time.sleep(1/f)
    p2.write(m1)
  else:
    break
p2.release()
```
