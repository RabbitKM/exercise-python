```python
# 動畫儲存
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time

p1 = cv2.VideoWriter('02.mp4', cv2.VideoWriter_fourcc(*'MP4V'), 10, (300,150))
for x in range(1,200,10):
  m1 = np.full((150, 300, 3), (0, 0, 0), np.uint8)
  cv2.circle(m1, (70,60), x, (0,255,0), 2)
  m1 = np.array(m1)

  dp.clear_output(True)
  cv2_imshow(m1)
  time.sleep(0.2)
  p1.write(m1)
p1.release()
```

```python
# 影片讀取
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time

p1 = cv2.VideoCapture("02.mp4")
while p1.isOpened()==True:
  ret, m1=p1.read()
  if ret==True:
    dp.clear_output(True)
    cv2.circle(m1, (20,20), 30, (0,255,0), -1)
    cv2_imshow(m1)
    time.sleep(0.3)
  else:
    break
```
