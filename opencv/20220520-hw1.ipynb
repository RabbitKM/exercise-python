```python
# 作業1
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time

while True:
  for x in range(0,301,5):
    m1 = np.full((200, 400, 3), (255, 255, 255), np.uint8)
    cv2.rectangle(m1, (x,50), (x+100,150), (255,0,0), -1)
    m1 = np.array(m1)

    dp.clear_output(True)
    cv2_imshow(m1)
    time.sleep(0.03)

  for y in range(300,-1,-5):
    m1 = np.full((200, 400, 3), (255, 255, 255), np.uint8)
    cv2.rectangle(m1, (y,50), (y+100,150), (255,0,0), -1)
    m1 = np.array(m1)

    dp.clear_output(True)
    cv2_imshow(m1)
    time.sleep(0.03)
```

```python
# 作業1-播放5次-存成動畫
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time

p1 = cv2.VideoWriter('hw1-2.mp4', cv2.VideoWriter_fourcc(*'MP4V'), 90, (400,200))
# while True:
for i in range(5):
  for x in range(0,301,2):
    m1 = np.full((200, 400, 3), (255, 255, 255), np.uint8)
    cv2.rectangle(m1, (x,50), (x+100,150), (255,0,0), -1)
    m1 = np.array(m1)

    dp.clear_output(True)
    cv2_imshow(m1)
    time.sleep(0.03)
    p1.write(m1)

  for y in range(300,-1,-2):
    m1 = np.full((200, 400, 3), (255, 255, 255), np.uint8)
    cv2.rectangle(m1, (y,50), (y+100,150), (255,0,0), -1)
    m1 = np.array(m1)

    dp.clear_output(True)
    cv2_imshow(m1)
    time.sleep(0.03)
    p1.write(m1)
p1.release()
```
