```python
# 第三解(去雜質) -過程去除
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

m1 = cv2.imread('h2.png', 1)

m2 = cv2.subtract(m1, (255,255,254,0)) # 只保留R=255
m3 = cv2.multiply(m2, (0,0,255,0)) # 還原R=255

m4 = cv2.bitwise_not(m3[:,:,2]) # 只取R+反色
m5 = cv2.morphologyEx(m4, cv2.MORPH_CLOSE, np.ones((2,1))) #去雜質
cv2_imshow(m5)
```

```python
# V第三解(去雜質-改簡短) 缺點: "夜"缺一小角
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

m1 = cv2.imread('h2.png', 1)

# 紅色保留1=R灰階-254
m2 = cv2.subtract(m1[:,:,2], 254)

# 轉回紅色=紅色1*255
m3 = cv2.multiply(m2, 255)

# 去雜質
m4 = cv2.morphologyEx(m3, cv2.MORPH_OPEN, np.ones((2,1)))

# 反色後輸出
cv2_imshow(cv2.bitwise_not(m4))
```

```python
# V第二解(改) - 圖較接近解答
import cv2
from google.colab.patches import cv2_imshow

m1 = cv2.imread('h2.png', 1)

# 紅色保留1=R灰階-254
m2 = cv2.subtract(m1[:,:,2], 254)

# 轉回紅色=紅色保留1*255
m3 = cv2.multiply(m2, 255)

# 去除G值 (為了排除: R=255 and G>0)
m4 = cv2.subtract(m3, m1[:,:,1])

# 去除B值 (為了排除: R=255 and B>0)
m5 = cv2.subtract(m4, m1[:,:,0])

# 反色後輸出
cv2_imshow(cv2.bitwise_not(m5))
```

```python
# 第二解(改2) - 答案一樣, 但處理順序感覺較合理.(先減去雜訊, 再*255)
import cv2
from google.colab.patches import cv2_imshow

m1 = cv2.imread('h2.png', 1)

# 紅色保留1=R灰階-254
m2 = cv2.subtract(m1[:,:,2], 254)

# 去除G有值 (為了排除: R=255 and G>0)
m3 = cv2.subtract(m2, m1[:,:,1])

# 去除B有值 (為了排除: R=255 and B>0)
m4 = cv2.subtract(m3, m1[:,:,0])

# 轉成白色=紅色保留1*255
m5 = cv2.multiply(m4, 255)

# 反色後輸出
cv2_imshow(cv2.bitwise_not(m5))
```
